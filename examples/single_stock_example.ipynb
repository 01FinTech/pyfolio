```python
# Here is a simple example where we produce a set of plots, called a tear sheet, for a single stock.
```

```python
%matplotlib inline
import pyfolio as pf
```

```python
stock_rets = pf.utils.get_symbol_rets('FB') # Fetch the daily returns for a stock
```

```python
# Create a full tear sheet for the single stock.
# This will show statistics and plots about returns and shock events.
# Set the first 80% of the stock's history to be in-sample (controls the cone plot).
# The Bayesian tear sheet is also enabled.
pf.create_full_tear_sheet(stock_rets, backtest_days_pct=0.8, bayesian=False)
```
